**Synthetic Data**

```python
data = np.random.randint(0, 2, size=(10, 2))

# Print the data
print(data)
```

```python
# Calculate the conditional probability P(Y=1|X=0)
numerator = np.sum((data[:, 1] == 1) & (data[:, 0] == 0))
denominator = np.sum(data[:, 0] == 0)
p_y_given_x = numerator / denominator
```

```python
# Print the result
print(f'P(Y=1|X=0) = {p_y_given_x:.2f}')
```

```python
import matplotlib.pyplot as plt
```

```python
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
```

```python
counts = pd.Series(data).value_counts()
```

```python
pmf = counts / len(data)
```

Overall, this code generates a simple bar plot of a probability mass function, which is a way of representing the probability distribution of a discrete random variable. The plot shows the probability of each possible value of the random variable on the y-axis, and the corresponding values on the x-axis.

```python
sns.barplot(x=pmf.index, y=pmf.values)
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Probability Mass Function')

# Show the plot
plt.show()
```

Overall, this code generates a simple step plot of the CDF of a dataset, which is a way of representing the probability distribution of a continuous random variable. The plot shows the cumulative probability of observing a value less than or equal to a given value on the x-axis.

```python
x, y = np.sort(data), np.arange(1, len(data)+1) / len(data)
plt.step(x, y)
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution Function')

# Show the plot
plt.show()
```

```python
mean, std = np.mean(data), np.std(data)
x = np.linspace(mean - 3*std, mean + 3*std, 100)
pdf = 1/(std * np.sqrt(2 * np.pi)) * np.exp(-(x - mean)**2 / (2 * std**2))

# Plot the PDF as a line plot
plt.plot(x, pdf)

# Add axis labels and title
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Probability Density Function')

# Show the plot
plt.show()
```

**binomial, Bernoulli, and Gaussian probability**

```python
# Binomial distribution
n = 10  # Number of trials
p = 0.5  # Probability of success
k = np.arange(0, n+1)  # Possible number of successes
binom_pmf = np.zeros_like(k, dtype=float)
for i in range(len(k)):
    binom_pmf[i] = np.math.comb(n, k[i]) * p**k[i] * (1-p)**(n-k[i])
plt.stem(k, binom_pmf)
plt.xlabel('Number of successes (k)')
plt.ylabel('Probability')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.show()
```

```python
# Bernoulli distribution
p = 0.5  # Probability of success
k = [0, 1]  # Possible outcomes
bernoulli_pmf = np.array([1-p, p])
plt.stem(k, bernoulli_pmf)
plt.xlabel('Outcome (k)')
plt.ylabel('Probability')
plt.title('Bernoulli Distribution (p=0.5)')
plt.show()
```

```python
# Gaussian distribution
mu = 0  # Mean
sigma = 1  # Standard deviation
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
gaussian_pdf = 1/(sigma * np.sqrt(2 * np.pi)) * np.exp(-(x - mu)**2 / (2 * sigma**2))
plt.plot(x, gaussian_pdf)
plt.xlabel('Value (x)')
plt.ylabel('Probability density')
plt.title('Gaussian Distribution (μ=0, σ=1)')
plt.show()
```
